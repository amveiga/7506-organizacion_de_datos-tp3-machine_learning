```python
import pandas as pd
```

```python
train_identity = pd.read_csv('/content/drive/MyDrive/2021-2C/ORGANIZACION DE DATOS (7506)/TPs/TP3/ieee-fraud-detection/train_identity.csv')
train_transaction = pd.read_csv('/content/drive/MyDrive/2021-2C/ORGANIZACION DE DATOS (7506)/TPs/TP3/ieee-fraud-detection/train_transaction.csv')

test_identity = pd.read_csv('/content/drive/MyDrive/2021-2C/ORGANIZACION DE DATOS (7506)/TPs/TP3/ieee-fraud-detection/test_identity.csv')
test_transaction = pd.read_csv('/content/drive/MyDrive/2021-2C/ORGANIZACION DE DATOS (7506)/TPs/TP3/ieee-fraud-detection/test_transaction.csv')
```

```python
joined_train = train_transaction.merge(train_identity,how='outer')

joined_test = test_transaction.merge(test_identity,how='outer')
```
